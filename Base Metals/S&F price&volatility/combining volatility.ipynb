```python
import numpy
import xlrd
import xlwt
```

```python
res = xlrd.open_workbook('result.xlsx')

volal = xlrd.open_workbook('Al volatility.xlsx')

volcopper = xlrd.open_workbook('Copper volatility.xlsx')

vollead = xlrd.open_workbook('Lead volatility.xlsx')

volnickel = xlrd.open_workbook('Nickel volatility.xlsx')

volTin = xlrd.open_workbook('Tin volatility.xlsx')

volZinc = xlrd.open_workbook('Zinc volatility.xlsx')
```

```python
tabler = res.sheets()[0]
tableal = volal.sheets()[0]
tablecopper = volcopper.sheets()[0]
tablelead = vollead.sheets()[0]
tablenickel = volnickel.sheets()[0]
tableTin = volTin.sheets()[0]
tableZinc = volZinc.sheets()[0]
```

```python
num_of_rows_res = tabler.nrows
```

```python
print(xlrd.xldate_as_tuple(tableal.cell(1,0).value, 0))
```

```python
output = {}
i = 0
j = 1

while i < num_of_rows_res and j < tableal.nrows:
    tup = xlrd.xldate_as_tuple(tabler.cell(i,0).value, 0)
    y = tup[0]
    m = tup[1]
    d = tup[2]
    print(y,m,d)
    while j < tableal.nrows:
        print(j)
        tupc = xlrd.xldate_as_tuple(tableal.cell(j,0).value, 0)
        print(tupc)
        yc = tupc[0]
        mc = tupc[1]
        print(yc,mc)
        if (yc == y and mc == m):
            date = str(y)+"/"+str(m)+"/"+str(d)
            output[date]=[tableal.cell(j,3).value, tablecopper.cell(j,3).value, tablelead.cell(j,3).value,
                          tablenickel.cell(j,3).value, tableTin.cell(j,3).value, tableZinc.cell(j,3).value]
            #print(j)
            #print(output[date])
            i = i +1
            break
        else:
            print(1)
            j += 1
```

```python
book = xlwt.Workbook()
sheet = book.add_sheet('cost')
row = 0

for d in output:
    sheet.write(row,0,d)
    sheet.write(row,1,output[d][0])
    sheet.write(row,2,output[d][1])
    sheet.write(row,3,output[d][2])
    sheet.write(row,4,output[d][3])
    sheet.write(row,5,output[d][4])
    sheet.write(row,6,output[d][5])

    row = row +1

book.save('ouput.xls')
```
