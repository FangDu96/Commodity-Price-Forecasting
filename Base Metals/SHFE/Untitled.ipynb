```python
import xlrd
import xlwt
import numpy as np
```

```python
rfbook = xlrd.open_workbook('China Bond Risk free rate -3m.xlsx')
table2012 = rfbook.sheets()[0]
table2013 = rfbook.sheets()[1]
table2014 = rfbook.sheets()[2]
table2015 = rfbook.sheets()[3]
table2016 = rfbook.sheets()[4]
table2017 = rfbook.sheets()[5]
table2018 = rfbook.sheets()[6]
```

```python
num12 = table2012.nrows
num13 = table2013.nrows
num14 = table2014.nrows
num15 = table2015.nrows
num16 = table2016.nrows
num17 = table2017.nrows
num18 = table2018.nrows
```

```python
i = 1
m = 1
ir12 = {}
ir12[m] = []
while i < num12:
    tup = xlrd.xldate_as_tuple(table2012.cell(i,0).value, 0)
    mi = tup[1]
    if mi == m:
        ir12[m].append(table2012.cell(i,2).value)
        i = i+1
    else:
        m = mi
        ir12[m] = []

        
i = 1
m = 1
ir13 = {}
ir13[m] = []
while i < num13:
    tup = xlrd.xldate_as_tuple(table2013.cell(i,0).value, 0)
    mi = tup[1]
    if mi == m:
        ir13[m].append(table2013.cell(i,2).value)
        i = i+1
    else:
        m = mi
        ir13[m] = []
        
i = 1
m = 1
ir14 = {}
ir14[m] = []
while i < num14:
    tup = xlrd.xldate_as_tuple(table2014.cell(i,0).value, 0)
    mi = tup[1]
    if mi == m:
        ir14[m].append(table2014.cell(i,2).value)
        i = i+1
    else:
        m = mi
        ir14[m] = []
        
i = 1
m = 1
ir15 = {}
ir15[m] = []
while i < num15:
    tup = xlrd.xldate_as_tuple(table2015.cell(i,0).value, 0)
    mi = tup[1]
    if mi == m:
        ir15[m].append(table2015.cell(i,2).value)
        i = i+1
    else:
        m = mi
        ir15[m] = []
        
i = 1
m = 1
ir16 = {}
ir16[m] = []
while i < num16:
    tup = xlrd.xldate_as_tuple(table2016.cell(i,0).value, 0)
    mi = tup[1]
    if mi == m:
        ir16[m].append(table2016.cell(i,2).value)
        i = i+1
    else:
        m = mi
        ir16[m] = []
        
i = 1
m = 1
ir17 = {}
ir17[m] = []
while i < num17:
    tup = xlrd.xldate_as_tuple(table2017.cell(i,0).value, 0)
    mi = tup[1]
    if mi == m:
        ir17[m].append(float(table2017.cell(i,2).value))
        i = i+1
    else:
        m = mi
        ir17[m] = []
        
i = 1
m = 1
ir18 = {}
ir18[m] = []
while i < num18:
    tup = xlrd.xldate_as_tuple(table2018.cell(i,0).value, 0)
    mi = tup[1]
    if mi == m:
        ir18[m].append(table2018.cell(i,2).value)
        i = i+1
    else:
        m = mi
        ir18[m] = []
```

```python
book = xlwt.Workbook()
sheet12 = book.add_sheet('2012')
sheet13 = book.add_sheet('2013')
sheet14 = book.add_sheet('2014')
sheet15 = book.add_sheet('2015')
sheet16 = book.add_sheet('2016')
sheet17 = book.add_sheet('2017')
sheet18 = book.add_sheet('2018')
row = 0
for m in ir12:
    sheet12.write(row,0,row+1)
    sheet12.write(row,1,np.mean(ir12[m]))
    row = row +1
    
row = 0
for m in ir13:
    sheet13.write(row,0,row+1)
    sheet13.write(row,1,np.mean(ir13[m]))
    row = row +1
    
row = 0
for m in ir14:
    sheet14.write(row,0,row+1)
    sheet14.write(row,1,np.mean(ir14[m]))
    row = row +1
    
row = 0
for m in ir15:
    sheet15.write(row,0,row+1)
    sheet15.write(row,1,np.mean(ir15[m]))
    row = row +1
    
row = 0
for m in ir16:
    sheet16.write(row,0,row+1)
    sheet16.write(row,1,np.mean(ir16[m]))
    row = row +1
    
row = 0
for m in ir17:
    sheet17.write(row,0,row+1)
    sheet17.write(row,1,np.mean(ir17[m]))
    row = row +1
    
row = 0
for m in ir18:
    sheet18.write(row,0,row+1)
    sheet18.write(row,1,np.mean(ir18[m]))
    row = row +1
book.save('IR.xls')
```
