```python
import numpy
import xlrd
import xlwt
```

```python
cost = xlrd.open_workbook('cost.xls')
spotal = xlrd.open_workbook('al_spot.xlsx')
futuresal = xlrd.open_workbook('al_3_futures.xlsx')

spotcopper = xlrd.open_workbook('copper_spot.xlsx')
futurescopper = xlrd.open_workbook('copper_3_futures.xlsx')

spotlead = xlrd.open_workbook('lead_spot.xlsx')
futureslead = xlrd.open_workbook('lead_3_futures.xlsx')

spotNickel = xlrd.open_workbook('Nickel_spot.xlsx')
futuresNickel = xlrd.open_workbook('Nickel_3_futures.xlsx')

spotZinc = xlrd.open_workbook('Zinc_spot.xlsx')
futuresZinc = xlrd.open_workbook('ZInc_3_futures.xlsx')

spotTin = xlrd.open_workbook('Tin_spot.xlsx')
futuresTin = xlrd.open_workbook('Tin_3_futures.xlsx')
```

```python
tableals = spotal.sheets()[0]
tablealf = futuresal.sheets()[0]

tablecoppers = spotcopper.sheets()[0]
tablecopperf = futurescopper.sheets()[0]

tableleads = spotlead.sheets()[0]
tableleadf = futureslead.sheets()[0]

tableNickels = spotNickel.sheets()[0]
tableNickelf = futuresNickel.sheets()[0]

tableZincs = spotZinc.sheets()[0]
tableZincf = futuresZinc.sheets()[0]

tableTins = spotTin.sheets()[0]
tableTinf = futuresTin.sheets()[0]
```

```python
tablec = cost.sheets()[0]
```

```python
num_of_rows_spot = tableals.nrows
num_of_rows_futures = tablealf.nrows
```

```python
num_of_rows_cost = tablec.nrows
print(num_of_rows_cost)
result = {}
i = 0
j = 1
yc = xlrd.xldate_as_tuple(tablec.cell(j,0).value, 0)[0]
mc = xlrd.xldate_as_tuple(tablec.cell(j,0).value, 0)[1]
print(yc,mc)

while (i < num_of_rows_spot and j < num_of_rows_cost):
    tup = xlrd.xldate_as_tuple(tableals.cell(i,0).value, 0)
    y = tup[0]
    m = tup[1]
    d = tup[2]
    print(y,m,d)
    if (y == 2019 and m == 4):
        date = str(y)+"/"+str(m)+"/"+str(d)
        result[date]=[tableals.cell(i,1).value, tablealf.cell(i,1).value, 0 ,
                      tablecoppers.cell(i,1).value, tablecopperf.cell(i,1).value, 0 ,
                      tableleads.cell(i,1).value, tableleadf.cell(i,1).value, 0 ,
                      tableNickels.cell(i,1).value, tableNickelf.cell(i,1).value, 0 ,
                      tableZincs.cell(i,1).value, tableZincf.cell(i,1).value, 0 ,
                      tableTins.cell(i,1).value, tableTinf.cell(i,1).value, 0]
        i += 1
    elif (y == yc and mc == m):
        date = str(y)+"/"+str(m)+"/"+str(d)
        result[date]=[tableals.cell(i,1).value, tablealf.cell(i,1).value, tablec.cell(j,4).value ,
                      tablecoppers.cell(i,1).value, tablecopperf.cell(i,1).value, tablec.cell(j,1).value ,
                      tableleads.cell(i,1).value, tableleadf.cell(i,1).value, tablec.cell(j,2).value ,
                      tableNickels.cell(i,1).value, tableNickelf.cell(i,1).value,tablec.cell(j,3).value ,
                      tableZincs.cell(i,1).value, tableZincf.cell(i,1).value,tablec.cell(j,6).value,
                      tableTins.cell(i,1).value, tableTinf.cell(i,1).value, tablec.cell(j,5).value]
        i += 1
    else:
        j += 1
        yc = xlrd.xldate_as_tuple(tablec.cell(j,0).value, 0)[0]
        mc = xlrd.xldate_as_tuple(tablec.cell(j,0).value, 0)[1]

```

```python
book = xlwt.Workbook()
sheet = book.add_sheet('cost')
row = 2

for d in result:
    sheet.write(row,0,d)
    sheet.write(row,1,result[d][0])
    sheet.write(row,2,result[d][1])
    sheet.write(row,3,result[d][2])
    sheet.write(row,4,result[d][3])
    sheet.write(row,5,result[d][4])
    sheet.write(row,6,result[d][5])
    sheet.write(row,7,result[d][6])
    sheet.write(row,8,result[d][7])
    sheet.write(row,9,result[d][8])
    sheet.write(row,10,result[d][9])
    sheet.write(row,11,result[d][10])
    sheet.write(row,12,result[d][11])
    sheet.write(row,13,result[d][12])
    sheet.write(row,14,result[d][13])
    sheet.write(row,15,result[d][14])
    sheet.write(row,16,result[d][15])
    sheet.write(row,17,result[d][16])
    sheet.write(row,18,result[d][17])
    row = row +1

book.save('result.xls')
```
